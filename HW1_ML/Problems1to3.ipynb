در این قسمت توابع را تعریف کردیم که تابع اول مقدار پیش بینی را با دیتای ورودی به همراه مقدار تتای تعیین شده محاسبه میکند.
تابع دوم تابع هزینه است که طبق مقادیر پیش بینی شده و مقادیر واقعی هزینه را مقایسه میکند.
توابع بعدی گرادیان نزولی هستند که در اینجا به دلیل خواسته های سوال من برای هر درجه یک تابع جداگانه که قسمت های مشابه زیادی دارند را تعریف کرده ام.
در هر تابع علاوه بر اینکه در هر ایپاک با محاسبه هزینه مقدار تتا تغییر پیدا میکند مقادیر خطا برای داده های اموزش و تست که در ورودی تابع داده شده اند نیز محاسبه میشود.همچنین برای هر بعد از بردار تتا نیز تعویض مقادیر آن ذخیره میشود تا بعدا با توجه به جهت رشد آنها که در نمودار رسم میشود به اهمیت این ابعاد پی ببریم

```python
def hypothesis(X, theta):
    y1 = X*theta
    return np.sum(y1, axis=1)


def cost(X, y, theta):
    y1 = hypothesis(X, theta)
    return sum(np.sqrt((y1-y)**2))/(2*m)


def gradientDescent3(X, y, theta, alpha, epoch, test_set):
    J=[]
    k=0
    mse = np.zeros(epoch)
    mae = np.zeros(epoch)
    rmse = np.zeros(epoch)
    test_mse = np.zeros(epoch)
    test_mae = np.zeros(epoch)
    test_rmse = np.zeros(epoch)
    thetaE0 = np.zeros(epoch)
    thetaE1 = np.zeros(epoch)
    thetaE2 = np.zeros(epoch)
    thetaE3 = np.zeros(epoch)
    

    while k < epoch:
        y1 = hypothesis(X, theta)
        for c in range(0, len(X.columns)):
            theta[c] = theta[c] - alpha*sum((y1-y)* X.iloc[:, c])/m
            if c==0:
              thetaE0[k] = theta[c]
            if c==1:
              thetaE1[k] = theta[c]
            if c==2:
              thetaE2[k] = theta[c]
            if c==3:
              thetaE3[k] = theta[c]
        j = cost(X, y, theta)
        J.append(j)
        mse[k] = np.square(np.subtract(y1,y )).mean()
        rmse[k] = np.sqrt(np.square(np.subtract(y1,y )).mean())
        mae[k] = np.abs(np.subtract(y1,y )).mean()
        test_pred = hypothesis(test_set.drop(columns= 'y'), theta)
        test_y = test_set['y']
        test_mse[k] = np.square(np.subtract(test_pred,test_y )).mean()
        test_rmse[k] = np.sqrt(np.square(np.subtract(test_pred,test_y )).mean())
        test_mae[k] = np.abs(np.subtract(test_pred,test_y )).mean()
        k += 1
    return J, theta, mse, mae, rmse, test_mse, test_mae, test_rmse, thetaE0, thetaE1, thetaE2, thetaE3 

def gradientDescent5(X, y, theta, alpha, epoch, test_set):
    J=[]
    k=0
    mse = np.zeros(epoch)
    mae = np.zeros(epoch)
    rmse = np.zeros(epoch)
    thetaE0 = np.zeros(epoch)
    thetaE1 = np.zeros(epoch)
    thetaE2 = np.zeros(epoch)
    thetaE3 = np.zeros(epoch)
    thetaE4 = np.zeros(epoch)
    thetaE5 = np.zeros(epoch)
    test_mse = np.zeros(epoch)
    test_mae = np.zeros(epoch)
    test_rmse = np.zeros(epoch)
    while k < epoch:
        y1 = hypothesis(X, theta)
        for c in range(0, len(X.columns)):
            theta[c] = theta[c] - alpha*sum((y1-y)* X.iloc[:, c])/m
            if c==0:
              thetaE0[k] = theta[c]
            if c==1:
              thetaE1[k] = theta[c]
            if c==2:
              thetaE2[k] = theta[c]
            if c==3:
              thetaE3[k] = theta[c]
            if c==4:
              thetaE4[k] = theta[c]
            if c==5:
              thetaE5[k] = theta[c]
        j = cost(X, y, theta)
        J.append(j)
        mse[k] = np.square(np.subtract(y1,y )).mean()
        rmse[k] = np.sqrt(np.square(np.subtract(y1,y )).mean())
        mae[k] = np.abs(np.subtract(y1,y )).mean()
        test_pred = hypothesis(test_set.drop(columns= 'y'), theta)
        test_y = test_set['y']
        test_mse[k] = np.square(np.subtract(test_pred,test_y )).mean()
        test_rmse[k] = np.sqrt(np.square(np.subtract(test_pred,test_y )).mean())
        test_mae[k] = np.abs(np.subtract(test_pred,test_y )).mean()
        k += 1
    return J, theta, mse, mae, rmse, test_mse, test_mae, test_rmse,thetaE0, thetaE1, thetaE2, thetaE3, thetaE4, thetaE5

def gradientDescent7(X, y, theta, alpha, epoch, test_set):
    J=[]
    k=0
    mse = np.zeros(epoch)
    mae = np.zeros(epoch)
    rmse = np.zeros(epoch)
    test_mse = np.zeros(epoch)
    test_mae = np.zeros(epoch)
    test_rmse = np.zeros(epoch)
    thetaE0 = np.zeros(epoch)
    thetaE1 = np.zeros(epoch)
    thetaE2 = np.zeros(epoch)
    thetaE3 = np.zeros(epoch)
    thetaE4 = np.zeros(epoch)
    thetaE5 = np.zeros(epoch)
    thetaE6 = np.zeros(epoch)
    thetaE7 = np.zeros(epoch)
    

    while k < epoch:
        y1 = hypothesis(X, theta)
        for c in range(0, len(X.columns)):
            theta[c] = theta[c] - alpha*sum((y1-y)* X.iloc[:, c])/m
            if c==0:
              thetaE0[k] = theta[c]
            if c==1:
              thetaE1[k] = theta[c]
            if c==2:
              thetaE2[k] = theta[c]
            if c==3:
              thetaE3[k] = theta[c]
            if c==4:
              thetaE4[k] = theta[c]
            if c==5:
              thetaE5[k] = theta[c]
            if c==6:
              thetaE6[k] = theta[c]
            if c==7:
              thetaE7[k] = theta[c]
        j = cost(X, y, theta)
        J.append(j)
        mse[k] = np.square(np.subtract(y1,y )).mean()
        rmse[k] = np.sqrt(np.square(np.subtract(y1,y )).mean())
        mae[k] = np.abs(np.subtract(y1,y )).mean()
        test_pred = hypothesis(test_set.drop(columns= 'y'), theta)
        test_y = test_set['y']
        test_mse[k] = np.square(np.subtract(test_pred,test_y )).mean()
        test_rmse[k] = np.sqrt(np.square(np.subtract(test_pred,test_y )).mean())
        test_mae[k] = np.abs(np.subtract(test_pred,test_y )).mean()
        k += 1
    return J, theta, mse, mae, rmse, test_mse, test_mae, test_rmse, thetaE0, thetaE1, thetaE2, thetaE3, thetaE4, thetaE5, thetaE6, thetaE7 
```

```python
def hypothesis(X, theta):
    y1 = X*theta
    return np.sum(y1, axis=1)


def cost(X, y, theta):
    y1 = hypothesis(X, theta)
    return sum(np.sqrt((y1-y)**2))/(2*m)


def gradientDescent(X, y, theta, alpha, epoch,test_set):
    J=[]
    k=0
    mse = np.zeros(epoch)
    mae = np.zeros(epoch)
    rmse = np.zeros(epoch)
    test_mse = np.zeros(epoch)
    test_mae = np.zeros(epoch)
    test_rmse = np.zeros(epoch)

    while k < epoch:
        y1 = hypothesis(X, theta)
        for c in range(0, len(X.columns)):
            theta[c] = theta[c] - alpha*sum((y1-y)* X.iloc[:, c])/m
            
        j = cost(X, y, theta)
        J.append(j)
        mse[k] = np.square(np.subtract(y1,y )).mean()
        rmse[k] = np.sqrt(np.square(np.subtract(y1,y )).mean())
        mae[k] = np.abs(np.subtract(y1,y )).mean()
        ytest = test_set['price']
        Xtest = test_set.drop(columns = 'price')
        y_hat = hypothesis(Xtest, theta)
        test_mse[k] = np.square(np.subtract(y_hat,ytest )).mean()
        test_rmse[k] = np.sqrt(np.square(np.subtract(y_hat,ytest )).mean())
        test_mae[k] = np.abs(np.subtract(y_hat,ytest )).mean()
        k += 1
    return J, theta, mse, mae, rmse, test_mse, test_mae, test_rmse
```
